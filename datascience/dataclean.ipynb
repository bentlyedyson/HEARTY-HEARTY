```python
import pandas as pd
from os import error
import numpy as np
```

```python
df_ptx = pd.read_csv(r'ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.1/ptbxl_database.csv')
df_scp = pd.read_csv(r'ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.1/scp_statements.csv')
```

```python
# df_ptx.isna().sum()
clean_ptx = df_ptx.drop([
    'infarction_stadium1',
    'infarction_stadium2',
    'nurse',
    'site',
    'device',
    'report',
    'recording_date',
    'validated_by',
    'second_opinion',
    'initial_autogenerated_report',
    'validated_by_human',
    'static_noise'
    ], axis =1)

clean_ptx.isna().sum()
```

```python

clean_ptx = clean_ptx[clean_ptx['electrodes_problems'].isnull()]
# clean_ptx['electrodes_problems'].isnull()
# clean_ptx.isna().sum()

clean_ptx = clean_ptx[clean_ptx['pacemaker'].isnull()]

clean_ptx = clean_ptx.drop([
    'electrodes_problems',
    'pacemaker'
    ],axis=1 )

clean_ptx.info()
```

```python
from sklearn.impute import KNNImputer

dummy_var = clean_ptx[['height','weight','age']]
dummy = pd.get_dummies(dummy_var, drop_first=True)
dummy_var.info()
dummy.info()

```

```python
clean_ptx = clean_ptx.drop(['height','weight','age'], axis = 1)
clean_ptx.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer
print(dummy.info())
scaler = MinMaxScaler()
dummy = pd.DataFrame(scaler.fit_transform(dummy), columns = dummy.columns, index = dummy.index)
print(dummy.info())
imputer = KNNImputer(n_neighbors=5)
dummy = pd.DataFrame(imputer.fit_transform(dummy),columns = dummy.columns, index = dummy.index)
print(dummy.info())
```

```python
dummy.info() 
```

```python
dummy_trans = scaler.inverse_transform(dummy)
dummy = pd.DataFrame({'height' : dummy_trans[:, 0], 'weight' : dummy_trans[:, 1], 'age' : dummy_trans[:, 2]}, columns = dummy.columns, index = dummy.index )
clean_ptx = pd.concat([clean_ptx, dummy], axis = 1)
```
