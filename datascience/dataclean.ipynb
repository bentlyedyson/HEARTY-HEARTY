```python
import pandas as pd
from os import error
import numpy as np
```

```python
df_ptx = pd.read_csv(r'datascience/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.1/ptbxl_database.csv')
df_scp = pd.read_csv(r'datascience/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.1/scp_statements.csv')
```

```python
# df_ptx.isna().sum()
clean_ptx = df_ptx.drop([
    'infarction_stadium1',
    'infarction_stadium2',
    'nurse',
    'site',
    'device',
    'report',
    'recording_date',
    'validated_by',
    'second_opinion',
    'initial_autogenerated_report',
    'validated_by_human',
    'static_noise'
    ], axis =1)

clean_ptx.isna().sum()
```

```python

clean_ptx = clean_ptx[clean_ptx['electrodes_problems'].isnull()]
# clean_ptx['electrodes_problems'].isnull()
# clean_ptx.isna().sum()

clean_ptx = clean_ptx[clean_ptx['pacemaker'].isnull()]

clean_ptx = clean_ptx.drop([
    'electrodes_problems',
    'pacemaker'
    ],axis=1 )

clean_ptx.info()
```

```python
from sklearn.impute import KNNImputer

other_dummy = clean_ptx[[
    'ecg_id', 
    'scp_codes',
    'heart_axis',
    'baseline_drift',
    'burst_noise',
    'extra_beats',
    'filename_lr',
    'filename_hr',]]
dummmy_var = clean_ptx[['height','weight']]
dummy = pd.get_dummies(dummmy_var, drop_first=True)
dummy.head()

```

```python
clean_ptx = clean_ptx.drop(['height', 'weight'],axis = 1)
clean_ptx = pd.concat([clean_ptx, dummy], axis = 1)
clean_ptx.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
clean_ptx = pd.DataFrame(scaler.fit_transform(clean_ptx), columns = clean_ptx.columns)
clean_ptx.head()
```
